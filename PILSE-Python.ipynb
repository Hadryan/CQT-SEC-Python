# Pitch-Invariant Log Spectral Envelope

## 01. Assumption and observations

**Assumption**: a log-spectrum, such as the CQT-spectrum, can be represented as to the convolution of a pitch-invariant log-spectal envelope component (~ timbre) and a envelope-normalized pitch component.

**Observation 1**: A pitch change in audio translates to a linear shift in the log-spectrum. 

This means that the same musical object at different pitches should have a similar envelope component but a shifted pitch component, while two different musical objects at the same pitch should have different envelope components but a similar pitch component

`X = E * P`

`X2 = E * P2`

```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load an audio file with its sampling frequency, and keep only one note in its left channel (= C4)
file_path = r"94812__digifishmusic__c4-major-scale-piano.wav"
audio_signal, sampling_frequency = librosa.load(file_path, sr=None, mono=False)
audio_signal = audio_signal[0, 24000:69000]

# Shift down the note by a whole octave (2x lower in frequency) (= C3)
audio_signal2 = librosa.effects.pitch_shift(audio_signal, sampling_frequency, n_steps=-12)

# Comptute the CQT-spectrogram for the both signals
step_length = int(pow(2, int(np.ceil(np.log2(0.04*sampling_frequency))))/2)
audio_spectrogram = np.abs(librosa.cqt(audio_signal, sr=sampling_frequency, hop_length=step_length))
audio_spectrogram2 = np.abs(librosa.cqt(audio_signal2, sr=sampling_frequency, hop_length=step_length))

# Display their waveform in seconds and power CQT-spectrogram in seconds and notes
plt.figure(figsize=(17, 7))
plt.subplot(2, 2, 1)
librosa.display.waveplot(audio_signal, sr=sampling_frequency)
plt.title("Original signal (C4)")
plt.subplot(2, 2, 2)
librosa.display.waveplot(audio_signal2, sr=sampling_frequency)
plt.title("Pitch-shifted signal (C3)")
plt.subplot(2, 2, 3)
librosa.display.specshow(librosa.amplitude_to_db(audio_spectrogram, ref=np.max), \
                         sr=sampling_frequency, hop_length=step_length, x_axis='time', y_axis='cqt_note')
plt.subplot(2, 2, 4)
librosa.display.specshow(librosa.amplitude_to_db(audio_spectrogram2, ref=np.max), \
                         sr=sampling_frequency, hop_length=step_length, x_axis='time', y_axis='cqt_note')
plt.show()
```

**Observation 2**: The Fourier transform (FT) of a convolution is equal to the product of the FTs (convolution theorem).

`X = E * P => F(X) = F(E) x F(P)`

`X2 = E * P2 => F(X2) = F(E) x F(P2)`

```python
import numpy as np
import matplotlib.pyplot as plt

# Define two signals
number_elements = 100
data_signal1 = np.random.normal(size=number_elements)
data_signal2 = np.random.normal(size=number_elements)

# Compute the direct convolution between them
data_convolution = np.convolve(data_signal1, data_signal2)

# Compute the convolution using the convolution theorem
ft_signal1 = np.fft.fft(data_signal1, 2*number_elements-1)
ft_signal2 = np.fft.fft(data_signal2, 2*number_elements-1)
data_convolution2 = np.real(np.fft.ifft(ft_signal1*ft_signal2))

# Display the convolutions and their differences
maximum_value = max(abs(data_convolution))
plt.figure(figsize=(17, 2))
plt.subplot(1, 3, 1)
plt.plot(data_convolution), plt.ylim(-maximum_value, maximum_value), plt.title("Direct convolution")
plt.subplot(1, 3, 2)
plt.plot(data_convolution2), plt.ylim(-maximum_value, maximum_value), plt.title("Indirect convolution")
plt.subplot(1, 3, 3)
plt.plot(data_convolution-data_convolution2), plt.ylim(-maximum_value, maximum_value), plt.title("Differences")
plt.show()
```

**Observation 3**: The magnitude FT is shift-invariant

`F(X) = abs(F(X)) x exp(j*angle(F(X)))`

`F(X2) = abs(F(X2)) x exp(j*angle(F(X2)))`

`abs(F(X)) ~ abs(F(X2)) => F(E) ~ abs(F(X)) => E ~ F-1(abs(F(X))) and P ~ F-1(exp(j*angle(F(X))))`

```python
import numpy as np
import matplotlib.pyplot as plt

# Define an array and a shifted version
N = 100
D = 10
x1 = np.random.normal(size=N)
x2 = np.concatenate((np.zeros(D),x1[0:N-D]))
M = max(np.max(abs(x1)), np.max(abs(x2)))
x1 = x1/M
x2 = x2/M

# Compute the magnitude FT of the arrays
X1 = np.fft.fft(x1)
X2 = np.fft.fft(x2)
A1 = abs(X1)
A2 = abs(X2)
P1 = np.angle(X1)
P2 = np.angle(X2)

# Display the magnitude FTs and their differences
M = max(np.max(A1), np.max(A2))
plt.figure(figsize=(17, 6))
plt.subplot(3, 3, 1), plt.plot(x1), plt.ylim(-1, 1), plt.title("Array")
plt.subplot(3, 3, 2), plt.plot(x2), plt.ylim(-1, 1), plt.title("Shifted array")
plt.subplot(3, 3, 4), plt.plot(A1), plt.ylim(0, M), plt.title("Magnitude FT of array")
plt.subplot(3, 3, 5), plt.plot(A2), plt.ylim(0, M), plt.title("Magnitude FT of shifted array")
plt.subplot(3, 3, 6), plt.plot(abs(A1-A2)), plt.ylim(0, M), plt.title("Differences")
plt.subplot(3, 3, 7), plt.plot(P1), plt.ylim(-np.pi, np.pi), plt.title("Angle FT of array")
plt.subplot(3, 3, 8), plt.plot(P2), plt.ylim(-np.pi, np.pi), plt.title("Angle FT of shifted array")
plt.show()
```

## 01. Read an audio file and display its magnitude CQT-spectrogram

```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load an audio file with its sampling frequency, and keep only its left channel
file_path = r"94812__digifishmusic__c4-major-scale-piano.wav"
audio_signal, sampling_frequency = librosa.load(file_path, sr=None, mono=False)
audio_signal = audio_signal[0, :]

# Comptute the magnitude CQT-spectrogram
step_length = int(pow(2, int(np.ceil(np.log2(0.04*sampling_frequency))))/2)
audio_spectrogram = np.abs(librosa.cqt(audio_signal, sr=sampling_frequency, hop_length=step_length))
number_frequencies, number_times = np.shape(audio_spectrogram)

# Display the power CQT-spectrogram in seconds and notes
plt.figure(figsize=(17, 10))
librosa.display.specshow(librosa.amplitude_to_db(audio_spectrogram, ref=np.max), sr=sampling_frequency, \
                         hop_length=step_length, x_axis='time', y_axis='cqt_note')
plt.title("CQT-spectrogram")
plt.show()
```

```python
np.shape(audio_spectrogram)
```

## 02. Compute the FT of the log spectra and derive the pitch and envelope components

```python
# Compute the FT of the columns of the magnitude log-spectrogram
ft_spectrogram = np.fft.fft(audio_spectrogram, 2*number_frequencies-1, axis=0)

# Decompose the results into amplitude and phase components
amplitude_component = abs(ft_spectrogram)
phase_component = ft_spectrogram/amplitude_component

# Derive the envelope and the pitch components
envelope_component = abs(np.fft.ifft(amplitude_component, axis=0))[0:number_frequencies, :]
pitch_component = abs(np.fft.ifft(phase_component, axis=0))[0:number_frequencies, :]

# Display the envelope and pitch components
plt.figure(figsize=(17, 10))
librosa.display.specshow(librosa.amplitude_to_db(envelope_component, ref=np.max), sr=sampling_frequency, \
                         hop_length=step_length, x_axis='time')
plt.title("CQT-envelope")
plt.show()
plt.figure(figsize=(17, 10))
librosa.display.specshow(pitch_component, sr=sampling_frequency, \
                         hop_length=step_length, x_axis='time', y_axis='cqt_note')
plt.title("CQT-pitch")
plt.show()
```

```python
audio_spectrogram2 = np.zeros((number_frequencies, number_times))
for i in range(number_times):
    audio_spectrogram2[:, i] = np.convolve(envelope_component[:, i], pitch_component[:, i])[0:number_frequencies]

plt.figure(figsize=(17, 10))
librosa.display.specshow(librosa.amplitude_to_db(audio_spectrogram2, ref=np.max), sr=sampling_frequency, \
                         hop_length=step_length, x_axis='time', y_axis='cqt_note')
plt.title("Resynthesized CQT-spectrogram")
plt.show()
```
